The p(t) used in this case was triangular

```python
from pylab import *
import ascfun as af
import wavfun as wf
fs, rt  = wf.wavread("pamsig201.wav")
fb= 251
tb = 1/float(fb)
bits = 8
n = int(floor(len(rt)/float(fs)/tb))  #number of recieved bits
rt= list(rt)

######### getting sample of rt signal #######
dnhat=[]
for i in range(n):                      # getting sample of rt signal
	dn_sample = rt[i*round(fs*tb):int(((i+1))*round(fs*tb))]
	avg = sum(dn_sample) / round(fs*tb) # averaging out the one bit window and the comapring
	if avg > 0.5:                       # Quantisation of the bits
		dnhat = dnhat + [1]
	else:
		dnhat = dnhat + [0]
#####################################################
dnhat = array(dnhat,int8)	 # converting list into binary array

print("")
print('The content of the wav file is "%s"' %af.bin2asc(dnhat))
print("")
```
