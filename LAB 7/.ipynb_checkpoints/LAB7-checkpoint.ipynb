### AMsig701 
#### By narrowing in the frequency scale it seems more like a SC signal the closer I look as their is no amplitude of the signal present at the fc.

#### Then I narrowed in on the frequency scale to estimate the Fc, in this case it appears to be around 9350. From the second psd I can tell that the fm is approximately equal to 3350. 

#### The first rtype I tried is abs, which is not able to extract the signal clearly but i can tell that it is a music signal. For rtype of iqabs the signal was cleaner but not enough. For rtype of coh with fc = 9350 and theta = 0, I get the cleanest signal. It is a music signal

#### However I was not able to demodulated the signal without any distortion. My guess is that my Fc or my thetac is not exactly correct.
```python
from pylab import *
import matplotlib.pyplot as plt
import wavfun
import comsig
import amfun
import showfun1

Fs, rt = wavfun.wavread('amsig701.wav')
sig_rt = comsig.sigWave(rt, Fs)
plt.plot(sig_rt.timeAxis()[0:1000], sig_rt.signal()[0:1000])
showfun1.showft(sig_rt.timeAxis(), sig_rt, Fs,[-3000,8000,0])
showfun1.showpsd(sig_rt, Fs, [-20000,20000,0], sig_rt.Nsamp)
showfun1.showpsd(sig_rt, Fs, [5000,14000,0], sig_rt.Nsamp)
showfun1.showpsd(sig_rt, Fs, [8500,9500,0], sig_rt.Nsamp)

ftype = 'LPF'
fcparms1 =[9350,(-90*np.pi)/180]
fcparms2 =[9350, 0]
fmparms  =[3350,10,0.05]
sig_mthat = amfun.amrcvr(sig_rt.timeAxis(),sig_rt,'iqangle',fcparms1,fmparms,[],ftype)
wavfun.wavwrite('iqar701.wav', Fs, 0.9999*sig_mthat.signal()/float(max(abs(sig_mthat.signal()))))






```

#### To demodulate the file AMsignal_005.bin I have built a flowgraph in the GRC for tuning to and demodulating AM-DSB-SC and, more generally QAM signals (i.e., the sum of two AM-DSB-SC signals at the same carrier frequency, one with a cosine and one with a sine carrier). # Expt 3(a)

### For Amsig708

#### From the PSD I can tell that it is a TC signal since the carrier has double the amplitude of the sidebands. I use the iqabs reciever because the phase is particularly difficult to determine. I am able to get a clean demodulated speech signal

```python
from pylab import *
import matplotlib.pyplot as plt
import wavfun
import comsig
import amfun
import showfun1

Fs, rt = wavfun.wavread('amsig708.wav')
sig_rt = comsig.sigWave(rt, Fs)
plt.plot(sig_rt.timeAxis()[0:500], sig_rt.signal()[0:500])
showfun1.showpsd(sig_rt, Fs, [-20000,20000,0], sig_rt.Nsamp)
#showfun1.showft(sig_rt.timeAxis(), sig_rt, Fs, [-20000,20000,-60])
#showfun1.showpsd(sig_rt, Fs, [5000,9000,-60], sig_rt.Nsamp)
#showfun1.showpsd(sig_rt, Fs, [13000,16000,-60], sig_rt.Nsamp)
ftype = 'LPF'
fcparms1 =[10000,(-90*np.pi)/180]
fcparms2 =[10000, 0]
fmparms  =[5000,10,0.05]
asig_mthat = amfun.amrcvr(sig_rt.timeAxis(),sig_rt,'iqabs',fcparms1,fmparms,[],ftype)
wavfun.wavwrite('iqabsr708.wav', Fs, 0.9999*sig_mthat.signal()/float(max(abs(sig_mthat.signal()))))
```
