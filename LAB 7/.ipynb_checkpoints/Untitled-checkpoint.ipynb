```python
import filtfun
import amfun
import showfun1
import numpy as np
import comsig
import wavfun

Fs = 44100 # Sampling rate 
# speech and music signal for modulating
FB1, mt1 = wavfun.wavread('speech701.wav')
sig_mt1 = comsig.sigWave(mt1, Fs, 0) # Waveform from class sigWave
xtype = 'sc'
fcparms1 =[8000,(-90*np.pi)/180]
fcparms2 =[8000, 0]
fmparms  =[4000,10,0.05]

#Transmitting the modulated signal
sig_xt1 = amfun.amxmtr(sig_mt1.timeAxis(),sig_mt1.signal(),xtype,fcparms1,fmparms)

#Show psd for the unmodulated signal
#showfun1.showpsd(sig_mt1, Fs, [-18000,18000, 0],sig_mt1.Nsamp)
#showpsd for the modulated signal
#showfun1.showpsd(sig_xt1, Fs, [4000,12000, 0],sig_xt1.Nsamp)

# Demodulating the signal
sig_mthat = amfun.amrcvr(sig_xt1.timeAxis(),sig_xt1,'coh',fcparms1, fmparms,[],'LPF')
#showpsd for the demodulated signal
#showfun1.showpsd(sig_mthat, Fs, [-18000,18000, 0],sig_mthat.Nsamp)
#wavfun.wavwrite('music701rec.wav',Fs,(0.999*sig_mthat.signal()/float(max(abs(sig_mthat.signal())))))
```
